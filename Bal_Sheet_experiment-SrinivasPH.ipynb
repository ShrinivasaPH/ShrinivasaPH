```python
print()
#Cash in Bank
city_bank = float(448430.24)
hdfc_bank = float(68615.52)
#cash in hand
cash_in_hand = float(1000)

cash_in_bank = city_bank + hdfc_bank
print("Summary:")
print()
print("Assets")
print("Cash in Bank:")
print(round(cash_in_bank,2))
print("Cash in hand: ")
print(round(cash_in_hand,2))
total_cash = cash_in_bank + cash_in_hand
print("Total Cash:",round(total_cash,2))
print()

#Wallets
sodexo_zeta = 0 #float(2763.95)
amazon_pay = float(272)
bns_wallet = 348.62
airTM = 1278.73
UPIlite = 0
wallets = sodexo_zeta + amazon_pay + bns_wallet + airTM + UPIlite
#investments

hdfc_security = float(10521.4) #Enter HDFC share value here ðŸ“ˆðŸ“‰ðŸ“ˆðŸ“‰ðŸ“ˆðŸ“‰ðŸ“ˆðŸ“‰ðŸ“ˆðŸ“‰
print("HDFC Security",hdfc_security)

#providentFund
employeeShare = 0 #float(211301)
employerShare = 0 #float(82202)
provident_fund = employeeShare + employerShare
print("Provident Fund:",provident_fund)
pension_contribution = 0  #float(86116)

gold_price_per_gram = 7476.60 #24 Carat #Enter here: âœ¨âœ¨âœ¨âœ¨âœ¨âœ¨

crypto_curr = float(1668.70) #Enter here: ðŸ’¹ðŸ’°ðŸ’±
gpay_goldVault = float(gold_price_per_gram * 1.1669)

SGB = float(gold_price_per_gram *2)   #âœ¨âœ¨âœ¨âœ¨âœ¨âœ¨âœ¨âœ¨âœ¨âœ¨âœ¨âœ¨âœ¨
investments = hdfc_security + provident_fund + crypto_curr + gpay_goldVault + SGB + pension_contribution
print("Investments:",round(investments,2))

#Debtors
debtor_mohan = 0
debtor_vijay = 0
debtors = debtor_mohan + debtor_vijay
print()

if debtors > 0:
  print("Debtors:")
  print(debtors)
  print("____________________")
  print("____________________")
else:
  print("____________________")
  print()
  print("No Debtors")
  print("____________________")

print()
print()

#Liabilities
print("Liabilities:")
print("Creditors:")
paid_installments = 20
liquiEMIPaid = float(10583.33) * paid_installments    #âœ¨âœ¨âœ¨âœ¨âœ¨âœ¨âœ¨
liquiLoans = float(254000) - round(liquiEMIPaid)
print("(LiquiEMI paid:",round(liquiEMIPaid,2),")",(round((liquiEMIPaid/254000)*100)),"%")
HDFC_credit = float(0)
usha = float(0)
rajendra = 0 # float(18000) + float(7000) + float(20000) - float(30000) #BG + Adige + Haasige, Paid to Rajendra(negative balance) respectively
individualLoans =  usha + rajendra
print("liquiloans Balance:",liquiLoans, "(", round((liquiLoans/254000)*100),"%", ")" ,"Usha:",usha , "Rajendra:",rajendra)
print("HDFC Credit Card:",HDFC_credit)
creditors = liquiLoans + individualLoans + HDFC_credit

liabilities = creditors
print("____________________")
print("Your Liabilities value:")
print("Rs.",round(liabilities,2))
print("____________________")
print("____________________")

asset = (total_cash + investments + debtors + wallets) - liabilities
print("Your Gross Asset Value:","Rs.", round(asset + liabilities,2))
print("Your Net Asset Value:","Rs.",round(asset,2))
print("Your liquid asset:","Rs.",round(total_cash,2),"(Cash)")


print("____________________")
print("____________________")
print()
print()


print("Details:")
print("Shares:",hdfc_security)
print("PF :",provident_fund)
print("Pension Contribution:",pension_contribution)
print("Crypto:",crypto_curr)
print("             ")
print("Wallets:")
print("Sodexo:",sodexo_zeta)
print("Amazon Pay:",amazon_pay)
print("BNS Wallet:",bns_wallet)
print("             ")
print("             ")
print("GPay Gold Vault (1030 mgs):","Rs.",round(gpay_goldVault,2),"(Current value)")
print("At the time of investment(05-02-2023): Rs.5680")

gpay_goldVault_buyingTime = float(5680 + 201 + 1001) # Price of gold at the time of buying GpayGold
gold_Profit_loss = gpay_goldVault - gpay_goldVault_buyingTime
if gold_Profit_loss > 0:
    print("Gold Profit:","Rs.",round(gold_Profit_loss,2),"âœ¨","(",(round((gpay_goldVault/gpay_goldVault_buyingTime) * 100-100,2)),"% ^)")
else:
    print("Gold Loss:","Rs.",round(gold_Profit_loss,2),"âœ¨","(",(round((gpay_goldVault/gpay_goldVault_buyingTime) * 100-100,2)),"%)")

SGB_buying_time = 6213 # Price of gold at the time of buying SGB
SGB_Prof = (gold_price_per_gram * 2) - (SGB_buying_time * 2) # Profit in SGB
tot_gold_prof = round(gold_Profit_loss + SGB_Prof,2) # Total Gold Profit

print("SGB Profit:",round(SGB_Prof,2),"âœ¨", "(",(round((gold_price_per_gram/SGB_buying_time)*100-100,2)), "% ^)")
print("Total Gold profit:",tot_gold_prof,"âœ¨","^")
print("_____________")
print("_____________")
print("_____________")
```

```python
from dateutil.relativedelta import relativedelta
from datetime import date
import pandas as pd

# Assuming the following variables are defined:
# gpay_goldVault_buyingTime, gpay_goldVault, gold_Profit_loss

gold = pd.DataFrame({
    'Buying Price' : gpay_goldVault_buyingTime,
    'Current Price' : round(gpay_goldVault, 2),
    'Buy Date' : ['2023-02-05'],
    'Today' : [date.today()],
    'Profit' : round(gold_Profit_loss, 2),
    'Up%^ âœ¨' : round((gpay_goldVault / gpay_goldVault_buyingTime) * 100 - 100, 2),
    'Duration': ''
})

# Convert 'Buy Date' and 'Today' to datetime
gold["Buy Date"] = pd.to_datetime(gold['Buy Date'])
gold["Today"] = pd.to_datetime(gold['Today'])

# Calculate duration in years and months using relativedelta
gold['Duration'] = gold.apply(lambda row: relativedelta(row['Today'], row['Buy Date']), axis=1)

# Format the duration to show as "X years Y months" or "X months" if no years
gold['Duration'] = gold['Duration'].apply(lambda x: f"{x.years} years, {x.months} months" if x.years > 0 else f"{x.months} months")

gold
```

```python
# SGB Table

gold_SGB = pd.DataFrame({
    "Today" :  [date.today()],
    "Buying Price" : SGB_buying_time,
    "Current Price" : gold_price_per_gram,
    "Qty" : 2,
    "Total Bought" : SGB_buying_time * 2,
    "Total Current" : gold_price_per_gram * 2,
    "Up%^ âœ¨" : round((gold_price_per_gram/SGB_buying_time) *100-100,2)
})
gold_SGB["Today"] = pd.to_datetime(gold['Today'])
gold_SGB["Profit"] = gold_SGB["Total Current"] - gold_SGB["Total Bought"]
gold_SGB
```
